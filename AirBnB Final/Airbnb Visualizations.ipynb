```python
import pandas as pd
import numpy as np

df=pd.read_csv('C:/Users/Rohit/Documents/crymaar/Portfolio Projects/cleaned_Airbnb.csv')
```

```python
df.shape
df.columns
```

```python
import os
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Country',y='Price',data=df)
xt=plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='No.of Amenities',y='Price',data=df)
xt=plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(12,6))
sns.factorplot(x='Property Type',y='Price',data=df, color='m', estimator=np.median, size =4.5, aspect=1.35)
xt=plt.xticks(rotation=90)
```

```python
corr_data = df.corr()
fig = plt.figure(figsize=(20,15))
sns.heatmap(corr_data,square=True,annot=True)
plt.show()
```

```python
plt.figure(figsize=(2,2))
fig, ax=plt.subplots()
ax.scatter(df['Room Type'],df['Security Deposit'])
ax.set_title('Room Type vs Security Deposit')
ax.set_xlabel('Room Type')
ax.set_ylabel('Security Deposit')
```

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(12,6))
sns.factorplot(x='Country',y='Review Scores Value',data=df, color='m', estimator=np.median, size =4.5, aspect=1.35)
xt=plt.xticks(rotation=90)
```
